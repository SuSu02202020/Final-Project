```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os
```

```python
kdata = pd.read_csv('C:/Users/titus/Desktop/Final Project/Final-Project/Weather data.csv', header=0)
data = pd.read_csv('temp_data.csv', header=0)
```

```python
X = kdata['temp'].values
y = data['real_temp'].values
X = np.reshape(X,(61113, 1))

```

```python
plt.scatter(X,y)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model
```

```python
model.fit(X, y)
print(model)
```

```python
print('Weight coefficients: ', model.coef_)
print('y-axis intercept: ', model.intercept_) 
```

```python
predictions = model.predict(X)
print(f"True output: {y[0]}")
print(f"Predicted output: {predictions[0]}")
print(f"Prediction Error: {predictions[0]-y[0]}")
print(f"relative error: {(predictions[0]-y[0])/y[0]}")
predictions.shape
```

```python
pd.DataFrame({"Predicted": predictions, "Actual": y, "Error": predictions - y,"relative error":(predictions-y)/y})[["Predicted", "Actual", "Error","relative error"]]
```

```python
x_min = X.min()
x_max = X.max()
```

```python
y_min_actual = y.min()
y_max_actual = y.max()
```

```python
y_min_predicted = model.predict(x_min)
y_max_predicted = model.predict(x_max)
print(f"Actual Min Value: {y_min_actual}")
print(f"Predicted Min Value: {y_min_predicted}")
print(f"Actual Max Value: {y_max_actual}")
print(f"Predicted Max Value: {y_max_predicted}")
```
